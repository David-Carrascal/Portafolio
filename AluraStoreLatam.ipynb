# 2. Ventas por categoría

```python
for nombre, df in zip(nombres_tiendas, tiendas):
    print(f"\n{nombre}:")
    ventas_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
    ventas_categoria_format = ventas_categoria.apply(lambda x: f"${x:,.0f}")
    print(ventas_categoria_format.to_string())
```

```python
print("Tienda con mayor venta en una sola categoría:")
ventas_maximas_categoria = {}

for nombre, df in zip(nombres_tiendas, tiendas):
    suma_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
    max_categoria = suma_por_categoria.max()
    ventas_maximas_categoria[nombre] = max_categoria

mejor_tienda = max(ventas_maximas_categoria, key=ventas_maximas_categoria.get)
print(f"{mejor_tienda} con ${ventas_maximas_categoria[mejor_tienda]:,.2f} en su mejor categoría")
```

```python
for nombre, df in zip(nombres_tiendas, tiendas):
    suma_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
    plt.figure(figsize=(6, 2))  # un poco más ancha y alta que antes
    plt.bar(suma_categoria.index, suma_categoria.values)
    plt.title(f"Ventas por Categoría - {nombre}", fontsize=10)
    plt.ylabel("Monto en pesos", fontsize=8)
    plt.xlabel("Categoría", fontsize=8)
    plt.xticks(rotation=45, fontsize=6, ha='right')
    plt.yticks(fontsize=6)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
print("Calificación Promedio por Tienda:")
for nombre, df in zip(nombres_tiendas, tiendas):
    promedio = df['Calificación'].mean()
    print(f"{nombre}: {promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
print("Productos Más y Menos Vendidos (por cantidad de registros):")
for nombre, df in zip(nombres_tiendas, tiendas):
    producto_ventas = df['Producto'].value_counts()
    mas_vendido = producto_ventas.idxmax()
    menos_vendido = producto_ventas.idxmin()
    print(f"\n{nombre}:")
    print(f"Producto más vendido: {mas_vendido} ({producto_ventas.max()} ventas)")
    print(f"Producto menos vendido: {menos_vendido} ({producto_ventas.min()} venta)")
```

```python
print("Tienda con el producto más vendido (más repetido):")
mas_vendidos = {}

for nombre, df in zip(nombres_tiendas, tiendas):
    producto_ventas = df['Producto'].value_counts()
    max_ventas = producto_ventas.max()
    mas_vendidos[nombre] = max_ventas

mejor_tienda = max(mas_vendidos, key=mas_vendidos.get)
print(f"{mejor_tienda} con {mas_vendidos[mejor_tienda]} unidades del producto más vendido")
```

```python
mas_vendidos = {nombre: df['Producto'].value_counts().max() for nombre, df in zip(nombres_tiendas, tiendas)}

plt.bar(mas_vendidos.keys(), mas_vendidos.values())
plt.title("Producto más vendido por tienda (frecuencia)")
plt.ylabel("Unidades")
plt.xlabel("Tienda")
plt.show()
```

# 5. Envío promedio por tienda

```python
print("Costo de Envío Promedio por Tienda:")
for nombre, df in zip(nombres_tiendas, tiendas):
    costo_promedio = df['Costo de envío'].mean()
    print(f"{nombre}: ${costo_promedio:.2f}")
```

```python
print("Tienda con menor costo de envío promedio:")
envios = {}

for nombre, df in zip(nombres_tiendas, tiendas):
    promedio_envio = df['Costo de envío'].mean()
    envios[nombre] = promedio_envio

mejor_tienda = min(envios, key=envios.get)
print(f"{mejor_tienda} con costo de envío promedio de ${envios[mejor_tienda]:.2f}")
```

```python
envios = {nombre: df['Costo de envío'].mean() for nombre, df in zip(nombres_tiendas, tiendas)}

plt.bar(envios.keys(), envios.values())
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo ($)")
plt.xlabel("Tienda")
plt.show()
```
